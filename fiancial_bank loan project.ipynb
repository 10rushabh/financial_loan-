```python
MOnthly trends by issue date  for total funded amount 
 What this code does
Sorts the DataFrame by issue_date.

Extracts month-year in "%b %Y" format.

Groups by that month-year, sums loan_amount, and converts to millions.

Plots the data as both an area chart (fill_between) and a line chart (plot).

Labels each point with the amount.

Adds title, axis labels, and grid.
```

```python
monthly_funded = (
    df.sort_values('issue_date')
      .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y'))
      .groupby('month_name', sort=False)['loan_amount']
      .sum()
      .div(1000000)
      .reset_index(name='loan_amount_millions')
)

plt.figure(figsize=(10, 5))
plt.fill_between(monthly_funded['month_name'], monthly_funded['loan_amount_millions'],
                 color='black', alpha=0.5)
plt.plot(monthly_funded['month_name'], monthly_funded['loan_amount_millions'],
         color='blue', linewidth=2)

for idx, row in monthly_funded.iterrows():
    plt.text(idx, row['loan_amount_millions'] + 0.1, f"{row['loan_amount_millions']:.2f}",
             ha='center', va='bottom', fontsize=9, color='black')

plt.title('Total Funded Amount by Month', fontsize=14)
plt.xlabel('mounth')
plt.ylabel('Funded Amount (₹ Millions)')
plt.xticks(ticks=range(len(monthly_funded)), labels=monthly_funded['month_name'], rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
monthly_received = (
    df.sort_values('issue_date')
      .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y'))
      .groupby('month_name', sort=False)['total_payment']
      .sum()
      .div(1_000_000)
      .reset_index(name='receive_amount_millions')
)

plt.figure(figsize=(10, 5))
plt.fill_between(monthly_received['month_name'], monthly_received['receive_amount_millions'],
                 color='red', alpha=0.5)
plt.plot(monthly_received['month_name'], monthly_received['receive_amount_millions'],
         color='black', linewidth=2)

for idx, row in monthly_received.iterrows():
    plt.text(idx, row['receive_amount_millions'] + 0.1, f"{row['receive_amount_millions']:.2f}",
             ha='center', va='bottom', fontsize=9, color='black')

plt.title('Total Received Amount by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Received Amount (₹ Millions)')
plt.xticks(ticks=range(len(monthly_received)), labels=monthly_received['month_name'], rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Filter for bad loans using .isin()
bad_loans = df[df['loan_status'].isin(['Charged Off'])]

# Monthly applications for bad loans
monthly_bad_loans = (
    bad_loans
    .sort_values('issue_date')
    .assign(month_name=lambda x: x['issue_date'].dt.strftime('%b %Y'))
    .groupby('month_name', sort=False)['id']
    .count()
    .reset_index(name='loan_applications_count')
)

# Plot
plt.figure(figsize=(10, 5))
plt.fill_between(monthly_bad_loans['month_name'], monthly_bad_loans['loan_applications_count'],
                 color='orange', alpha=0.5)
plt.plot(monthly_bad_loans['month_name'], monthly_bad_loans['loan_applications_count'],
         color='darkorange', linewidth=2)

for i, row in monthly_bad_loans.iterrows():
    plt.text(i, row['loan_applications_count'] + 0.5,
             f"{row['loan_applications_count']}",
             ha='center', va='bottom', fontsize=9, rotation=0, color='black')

plt.title('Charged Off Loan Applications by Month', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Number of Applications')
plt.xticks(ticks=range(len(monthly_bad_loans)),
           labels=monthly_bad_loans['month_name'], rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
# Filter bad loans using .isin()
bad_loans = df[df['loan_status'].isin(['Charged Off'])]

# Group by state and sum loan amounts
state_funding = (
    bad_loans.groupby('address_state')['loan_amount']
    .sum()
    .sort_values(ascending=True)
)

# Convert to thousands
state_funding_thousands = state_funding / 1000

# Plot horizontal bar chart
plt.figure(figsize=(10, 8))
bars = plt.barh(state_funding_thousands.index,
                state_funding_thousands.values,
                color='lightcoral')

# Add labels to bars
for bar in bars:
    width = bar.get_width()
    plt.text(width + 10, bar.get_y() + bar.get_height() / 2,
             f"{width:,.0f}K", va='center', fontsize=9)

# Titles and labels
plt.title('Total Funded Amount by State for Charged Off Loans (in ₹ Thousands)')
plt.xlabel('Funded Amount (₹ \'000)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```
